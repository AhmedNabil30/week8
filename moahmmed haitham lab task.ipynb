```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def time_complexity_ubcf(n, m, k):
    # Time for calculating similarities: O(n^2 * m)
    similarity_time = n ** 2 * m

    # Time for generating predictions: O(n * k * m)
    prediction_time = n * k * m

    total_time = similarity_time + prediction_time
    return total_time

def time_complexity_ibcf(n, m, k):
    # Time for calculating similarities: O(m^2 * n)
    similarity_time = m ** 2 * n

    # Time for generating predictions: O(n * k * m)
    prediction_time = n * k * m

    total_time = similarity_time + prediction_time
    return total_time

def space_complexity_ubcf(n, m):
    # Space for user-item matrix: O(n * m)
    user_item_space = n * m

    # Space for similarity matrix: O(n^2)
    similarity_space = n ** 2

    total_space = user_item_space + similarity_space
    return total_space

def space_complexity_ibcf(n, m):
    # Space for user-item matrix: O(n * m)
    user_item_space = n * m

    # Space for similarity matrix: O(m^2)
    similarity_space = m ** 2

    total_space = user_item_space + similarity_space
    return total_space

def calculate_sparsity(user_item_matrix):
    # Total number of entries (n * m)
    total_entries = user_item_matrix.size

    # Count the number of non-zero entries in the matrix
    non_zero_entries = np.count_nonzero(user_item_matrix)

    # Calculate sparsity
    sparsity = 1 - (non_zero_entries / total_entries)

    return sparsity

# New user-item matrix (rows: users, columns: items)
user_item_matrix = np.array([
    [5, 3, 0, 0, 1],    # User 1
    [4, 0, 0, 2, 1],    # User 2
    [0, 0, 5, 4, 3],    # User 3
    [0, 2, 4, 5, 0],    # User 4
    [3, 0, 4, 0, 5]     # User 5
])

n_users, n_items = user_item_matrix.shape
k_neighbors = 2  # Number of nearest neighbors to consider for predictions

# Calculate Sparsity
sparsity = calculate_sparsity(user_item_matrix)

# Print a clean header for the results
print("="*50)
print(f"{'User-Item Matrix Analysis':^50}")
print("="*50)
print(f"Sparsity of the user-item matrix: {sparsity:.2f}")
print()

# Time Complexity Analysis
print("="*50)
print(f"{'Time Complexity Analysis':^50}")
print("="*50)
print(f"{'User-Based Collaborative Filtering (UBCF):':<40} {time_complexity_ubcf(n_users, n_items, k_neighbors):,} operations")
print(f"{'Item-Based Collaborative Filtering (IBCF):':<40} {time_complexity_ibcf(n_users, n_items, k_neighbors):,} operations")
print()

# Space Complexity Analysis
print("="*50)
print(f"{'Space Complexity Analysis':^50}")
print("="*50)
print(f"{'User-Based Collaborative Filtering (UBCF):':<40} {space_complexity_ubcf(n_users, n_items):,} units of memory")
print(f"{'Item-Based Collaborative Filtering (IBCF):':<40} {space_complexity_ibcf(n_users, n_items):,} units of memory")
print("="*50)

# Plotting the User-Item Matrix as a Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(user_item_matrix, annot=True, cmap="coolwarm", cbar_kws={'label': 'Rating'},
            xticklabels=[f'Item {i+1}' for i in range(n_items)],
            yticklabels=[f'User {i+1}' for i in range(n_users)], fmt='d', linewidths=0.5)

plt.title('User-Item Rating Matrix')
plt.xlabel('Items')
plt.ylabel('Users')
plt.tight_layout()

# Show the plot
plt.show()
```
